```python
!pip install streamlit
import pandas as pd
import streamlit as st

# Ваши данные (пример)
total = data

# Создание DataFrame
df = pd.DataFrame(total)

# Главная страница Streamlit
def main():
    st.title('Уровень опасности городов')

    # Таблица с данными
    st.write('Данные о городах:')
    st.table(df)

    # Выбор города
    selected_city = st.selectbox('Выберите город:', df['city'])

    # Отображение уровня опасности для выбранного города
    danger_level = df[df['city'] == selected_city]['danger_level'].values[0]
    st.write(f"Уровень опасности в городе {selected_city}: {danger_level}")

if __name__ == '__main__':
    main()
!streamlit run html.ipynb
```
